```python
with open(r'C:\Users\hbbel\OneDrive\Desktop\Research\Git Repos\Principle-Elicitation\Eval data\alert_test_100_dpo_comparison.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
print(data[0].keys())
output = []
for item in data:
    prompt = item['prompt_raw']
    prompt = prompt.replace('### Instruction:\n', '')
    prompt = prompt.replace('\n### Response:', '')


    icai = item['answer_dpo_icai_mistral']
    sacai = item['answer_dpo_ours_mistral']
    output.append({'prompt':prompt, 'icai':icai, 'scai':sacai})

with open('human_eval_prompts.json', 'w') as f:
    json.dump(output, f, indent=4)
```
